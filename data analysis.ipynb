```python
import pandas as pd
df = pd.read_csv("Wednesday.csv")
df['transit_timestamp'] = pd.to_datetime(df['transit_timestamp'])
morning_df = df[(df['transit_timestamp'].dt.time >= pd.Timestamp('07:00:00').time()) & 
                (df['transit_timestamp'].dt.time <= pd.Timestamp('09:00:00').time())]
```

```python
lines = {
    "Line1": ['R192', 'R190', 'R188', 'R186', 'R185', 'R183', 'R182', 'R180', 'R178', 'R176',
              'R174', 'R173', 'R172', 'R170', 'R168', 'R164', 'R162', 'R160', 'N049', 'R154',
              'N060', 'R135', 'R133', 'R131', 'R129', 'N512', 'R125', 'R123', 'R119', 'R117', 'R116', 'R106', 'R103', 'R101'],

    "Line2": ['R336', 'R335', 'R334', 'R333', 'R332', 'R331', 'R330', 'R329', 'R328', 'R326', 'R325', 'R323', 'R322', 'R320',
              'R318', 'R316', 'R314', 'R312', 'R310', 'R261', 'R306', 'R304', 'R302', 'R301', 'R168', 'R161B', 'N060', 'R135',
              'N512', 'N094', 'N095', 'R111', 'R600', 'R602', 'R606', 'R608', 'R610', 'R617', 'R619', 'R621', 'R622', 'R635',
              'R636', 'R637', 'R639', 'R641', 'R643', 'R645'],
    # "Line3": [...],  # 若有更多线路，继续这样添加
}

# 从所有线路中找出中转站
all_stations = [station for sublist in lines.values() for station in sublist]
transfer_stations = {station for station in all_stations if all_stations.count(station) > 1}
print(all_stations)
print(transfer_stations)
```

```python
# 初始化一个空的需求字典
demands = {}

# 计算时间点的平均流量
def calculate_avg_ridership(station, line):
    times = [pd.Timestamp('07:00:00').time(), pd.Timestamp('08:00:00').time(), pd.Timestamp('09:00:00').time()]
    total = 0
    for time in times:
        total += morning_df[(morning_df['station_complex_id'] == station) & 
                            (morning_df['transit_timestamp'].dt.time == time)]['ridership'].sum()
    return total / len(times)

# 遍历每一条线路
for line, stations in lines.items():
    previous_demand = 0
    for station in stations:
        avg_ridership = calculate_avg_ridership(station, line)
        
        # 如果站点是中转站，考虑所有线路的平均乘客流量
        if station in transfer_stations:
            total_avg_ridership = sum([calculate_avg_ridership(station, route) 
                                       for route in lines.keys() if station in lines[route]])
        else:
            total_avg_ridership = avg_ridership

        # 计算需求值
        demand = 0.6 * previous_demand + total_avg_ridership
        demands[station] = demand
        
        # 更新 previous_demand
        previous_demand = demand

# 将计算的需求值添加到morning_df中
morning_df['demand'] = morning_df['station_complex_id'].map(demands)

# 获取一二号线的站点
line1_and_line2_stations = lines["Line1"] + lines["Line2"]

# 仅过滤出一二号线的数据
line1_and_line2_df = morning_df[morning_df['station_complex_id'].isin(line1_and_line2_stations)]

# 显示结果
print(line1_and_line2_df[['station_complex_id', 'demand']].drop_duplicates())

```
